```python
headers = {
    "X-API-Key": "e968eae5-b3a2-3222-b161-b1ee524dc138",
    "X-API-Secret": "1056ca8cfc260879587cbcd01cedf5f3b22f38a3cafee291292716689e67f344"
}
API_BASE_URL = "https://swissdox.linguistik.uzh.ch/api"
API_URL_DOWNLOAD = f"{API_BASE_URL}/download/6a1d6909-36cc-449c-8a3c-168b24e81c84__2025_10_06T12_14_42.tsv.xz"
 
r = requests.get(
    API_URL_DOWNLOAD,
    headers=headers
)
if r.status_code == 200:
    print("Size of file: %.2f KB" % (len(r.content)/1024))
    fp = open("./swiss_prime_site.tsv.xz", "wb")
    fp.write(r.content)
    fp.close()
else:
    print(r.text)

with lzma.open("swiss_prime_site.tsv.xz") as f_in:
    content = f_in.read().decode("utf-8")


df = pd.read_csv(StringIO(content), sep="\t")

# --- helpers ---------------------------------------------------------------
ID_BLOCK = re.compile(r"<id>.*?</id>", flags=re.I | re.S)   # remove whole <id>...</id> block
PARA_OPEN = re.compile(r"<\s*p[^>]*>", flags=re.I)
PARA_CLOSE = re.compile(r"</\s*p\s*>", flags=re.I)
BR_TAG = re.compile(r"<\s*br\s*/?>", flags=re.I)
TAGS = re.compile(r"<[^>]+>")  # any remaining tag

def clean_html_text(x: object, keep_paragraphs=False) -> object:
    """Return plain text. Preserves paragraph breaks if keep_paragraphs=True."""
    if pd.isna(x):
        return pd.NA
    t = str(x)

    # 1) drop metadata blocks like <id>...</id>
    t = ID_BLOCK.sub(" ", t)

    # 2) convert paragraph/br to line breaks (optional)
    if keep_paragraphs:
        t = PARA_OPEN.sub("", t)
        t = PARA_CLOSE.sub("\n\n", t)
        t = BR_TAG.sub("\n", t)
    else:
        # collapse paragraphs to spaces
        t = PARA_OPEN.sub(" ", t)
        t = PARA_CLOSE.sub(" ", t)
        t = BR_TAG.sub(" ", t)

    # 3) strip any other tags: <tx>, <au>, <em>, …
    t = TAGS.sub(" ", t)

    # 4) decode HTML entities (e.g., &auml; → ä, &nbsp; → space)
    t = html.unescape(t)

    # 5) normalize whitespace
    t = re.sub(r"\s+", " ", t).strip() if not keep_paragraphs else \
        re.sub(r"[ \t]+", " ", t).strip()

    return t

# --- apply to your columns --------------------------------------------------
df["head_clean"] = df["head"].apply(clean_html_text)                 # single line
df["content_clean"] = df["content"].apply(lambda x: clean_html_text(x, keep_paragraphs=True))  # keeps paras


df_result = df[["pubtime","language", "char_count", "medium_name","head_clean","content_clean"]]


df_result.to_csv("swiss_prime_site.csv", index=False)
```
